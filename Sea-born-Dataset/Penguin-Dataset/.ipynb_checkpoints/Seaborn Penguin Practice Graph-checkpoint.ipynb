# PENGUIN DATASET FROM SEABORN EXAMPLE DATASETS

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguin = sns.load_dataset('penguins')
```

```python
penguin
```

## DATA CLEANING
**Remove useless characters. Extract relevant content from a Series. Check NaN values. Change the type of your Series**

```python
penguin.species.unique()
```

```python
penguin.island.unique()
```

No useless characters.

```python
penguin.isnull().sum()
```

**Text Columns**: Penguin sexes have a 11 NaN values. 

**Numerical Columns**: Each Body Mass, Penguin Bill length, Penguin Bill Depth length, Flipper length have 2 missing values.

```python
# Text Columns
# Dropping Null Values
penguin['sex'] = penguin['sex'].dropna()

# Numerical Columns
# Dropping Null Values
penguin['bill_length_mm'] = penguin['bill_length_mm'].dropna()
penguin.info()
```

```python
# Numerical Columns
# Replacing Null Values using mean
# penguin['bill_length_mm'] = penguin['bill_length_mm'].fillna(penguin['bill_length_mm'].mean())
# penguin['flipper_length_mm'] = penguin['flipper_length_mm'].fillna(penguin['flipper_length_mm'].mean())
# penguin['bill_depth_mm'] = penguin['bill_depth_mm'].fillna(penguin['bill_depth_mm'].mean())
```

```python
# Change the object columns as Categorical Data
penguin['species'] = pd.Categorical(penguin.species)
penguin['sex'] = pd.Categorical(penguin.sex)
penguin['island'] = pd.Categorical(penguin.island)

penguin.info()
```

```python
Adelie = penguin[penguin.species == 'Adelie']
Adelie
```

```python
Chinstrap = penguin[penguin.species == 'Chinstrap']
Chinstrap
```

```python
Gentoo = penguin[penguin.species == 'Gentoo']
Gentoo
```

## DATA VISUALIZATION

**Penguin Population and Gender Per Species** | Stacked Bar

```python
sns.set(style = "ticks")

penguin_sex = penguin.groupby(by = 'sex').size()

fig_1, ax = plt.subplots(figsize=(20, 10))

ax.bar(penguin.species, penguin_sex['Male'], color = '#90CAF9', label = 'Male')
ax.bar(penguin.species, penguin_sex['Female'], bottom = penguin_sex['Male'], color = '#EF5350', label = 'Female')

ax.set_title('Penguin Population and Gender Per Species', fontweight = 'bold', fontsize = 20) 
ax.set_xlabel('Species', fontsize = 15)
ax.set_ylabel('Population by Gender', fontsize = 15)
ax.legend()
```

Conclusion: Balanced Population between genders

**Species and Body Mass of a Penguin** | Box Plot

```python
sns.set_style('whitegrid')

fig_2, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))

fig_2.suptitle('Body mass difference between a female penguin and male penguin', y=1.05, fontweight = 'bold', fontsize = '25')

sns.boxplot(ax = axes[0], data = Adelie, x = 'body_mass_g', y = 'sex')
axes[0].set_title('Adelie Body Masses', fontweight = 'bold', fontsize = 20)
axes[0].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[0].set_ylabel('Penguin Sexes', fontsize = 12)

sns.boxplot(ax = axes[1], data = Chinstrap, x = 'body_mass_g', y = 'sex')
axes[1].set_title('Chinstrap Body Masses', fontweight = 'bold', fontsize = 20)
axes[1].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[1].set_ylabel('Penguin Sexes', fontsize = 12)

sns.boxplot(ax = axes[2], data = Gentoo, x = 'body_mass_g', y = 'sex')
axes[2].set_title('Gentoo Body Masses', fontweight = 'bold', fontsize = 20)
axes[2].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[2].set_ylabel('Penguin Sexes', fontsize = 12)
```

### Is there a difference between the female penguin and male penguin body mass?
    Assumption: Yes
    Conclusion: Yes, female penguin are generally lighter than the male penguin.
        Adelie Female Penguins are the lightest penguin and Gentoo Male Penguins are the heaviest.

### Do Flipper Length, Bill Length and Depth of a Penguin contribute to the Body Mass?

```python
sns.set(style = "whitegrid")

fig_3, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))
fig_3.suptitle('Relation of the Penguin Body Parts with Each other', fontweight = 'bold', fontsize = '20')

# Flipper Length and Bill Length
sns.scatterplot(ax = axes[0], data = penguin, x = 'flipper_length_mm', y = 'flipper_length_mm')
axes[0].set(xlabel = 'Flipper Length', ylabel = 'Bill Length')

# Flipper Length and Bill Depth
sns.scatterplot(ax = axes[1], data = penguin, x = 'flipper_length_mm', y = 'bill_depth_mm')
axes[1].set(xlabel = 'Flipper Length', ylabel = 'Bill Depth')

# Bill Depth and Bill Length
sns.scatterplot(ax = axes[2], data = penguin, x = 'bill_length_mm', y = 'bill_depth_mm')
axes[2].set(xlabel = 'Bill Length', ylabel = 'Bill Depth')
```

### Is there a relation between the body parts of the penguin?
    Assumption: Bill depth and Bill length are positively relates to each other
    Conclusion: 
        1. As the value of Flipper Length rise, the Bill Length also gets longer. We conclude that they are positively relates to each other. (ASSUMPTION 1: Maybe Penguin body needs the Flipper Length and Bill length proportional to maintain balance)
        2. The Flipper Length seems to stay two patterns in the range of 175-225 of the increasing value of the Bill Length. We conclude that Flipper Length don't significally affect the Bill depth regardless of its value.
        3. Same with 3, the Bill length maintain its value even the Bill Depth increases. We conclude that the Bill Depth don't significantly affect the Bill Length regardless of its value.

```python
sns.set_style('whitegrid')

fig_4, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))

sns.barplot(ax = axes[0], x = 'species', y = 'flipper_length_mm', data = penguin, ci = None, color ='#FFEB3B')
axes[0].set_title('Which Penguin Species has longer flipper?', fontweight = 'demibold', fontsize = 17)
axes[0].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[0].set_ylabel('Flipper Length (in mm)', fontsize = 12)

sns.barplot(ax = axes[1], x = 'species', y = 'bill_length_mm', data = penguin, ci = None, color ='#FFEB3B')
axes[1].set_title('Which Penguin Species has longer bill?', fontweight = 'demibold', fontsize = 17)
axes[1].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[1].set_ylabel('Bill Length (in mm)', fontsize = 12)

sns.barplot(ax = axes[2], x = 'species', y = 'bill_depth_mm', data = penguin, ci = None, color ='#FFEB3B')
axes[2].set_title('Which Penguin Species has larger bill?', fontweight = 'demibold', fontsize = 17)
axes[2].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[2].set_xlabel('Bill Depth (in mm)', fontsize = 12)
```

### What Penguin species have larger body parts?
    Gentoo Penguins have the longest flipper. Chinstrap Penguins have the longest bill. Adelie Penguins and Chinstrap Penguins both have the largest bill.

```python
sns.set(style = 'whitegrid')

fig_5, axes = plt.subplots(1, 2, sharex = True, figsize = (20, 5.75))

ax = sns.scatterplot(ax = axes[0], data = penguin, x = "flipper_length_mm", y = "bill_length_mm", size = "body_mass_g", color = '#194A8D',legend = False, sizes = (50, 100))
ax = sns.scatterplot(ax = axes[1], data = penguin, x = "flipper_length_mm", y = "bill_depth_mm", size = "body_mass_g", color = '#194A8D',legend = False, sizes = (50, 100))
plt.show()
```

## Penguin body parts affect its body mass
    Flipper Length and Bill Length relates to each other than Flipper Length and Bill Length. Longer Flipper and Bill adds on weight that why Gentoo Penguins have highest body mass followed by the Chinstrap Penguins. The Adelie is the lowest among the two category. 
    
    QUESTION FROM BODY MASS OF THE PENGUIN CHAPTER: Because the flipper length and bill length of the Chinstrap and Adelie are more closer to each other, they are more closer in body mass compared to the Gentoo Penguins. 
    
    We conclude that the Flipper Length and Bill are directly proportional to the body mass of the penguin to maintain its balance.

```python
sns.set_style('whitegrid')

fig_6, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))

sns.histplot(ax = axes[0], x = 'island', data = Adelie)
axes[0].set_title('Adelie Living Spaces', fontweight = 'demibold', fontsize = 17)
axes[0].set(xlabel='Living Spaces', ylabel='Count') 

sns.histplot(ax = axes[1], x = 'island', data = Chinstrap)
axes[1].set_title('Chinstrap Living Spaces', fontweight = 'demibold', fontsize = 17)
axes[1].set(xlabel='Living Spaces', ylabel='Count') 

sns.histplot(ax = axes[2], x = 'island', data = Gentoo)
axes[2].set_title('Gentoo Living Spaces', fontweight = 'demibold', fontsize = 17)
axes[2].set(xlabel='Living Spaces', ylabel='Count') 
```
