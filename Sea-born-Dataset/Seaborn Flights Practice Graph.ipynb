## DATA VISUALIZATION

```python
sns.set_style('whitegrid')

fig_1 = plt.figure(figsize=(15,8))

ax=sns.barplot(x = 'passengers', y = 'month', orient = 'h',data = flights, ci = None)
ax.set_title('Passenger Count every month', fontweight = 'demibold', fontsize = 17)
ax.set_xlabel('Passenger Count', fontsize = 12)
ax.set_ylabel('Month', fontsize = 12)
```

```python
sns.set_style('whitegrid')

fig_2 = plt.figure(figsize=(15,8))

ax=sns.boxplot(data = flights, x = 'month', y = 'passengers')
ax.set_title('Passenger Count every month', fontweight = 'demibold', fontsize = 17)
ax.set_xlabel('Passenger Count', fontsize = 12)
ax.set_ylabel('Month', fontsize = 12)
```

### Month and Passenger Count
    Conclusion: Passengers that have flights near the end and start of the year are lesser than the between months of the year. August and June are the highest and November is the lowest.
    
    Why? Assumptions
        1. Flights between the year are cheaper than starting or end of the year.
        2. June to August are the good times for vacations and business trips.
        3. Places that have flights near the end/start of the year are unpopular.
        4. Places that have flights near the end/start of the year have unpredictable/bad weather or prone to disasters.

```python
sns.set_style('whitegrid')

fig_3 = plt.figure(figsize=(15,8))

ax = sns.lineplot(x = "year", y = "passengers", data = flights, marker='s') 
ax.set_title('Passenger Count every year', fontweight = 'demibold', fontsize = 17)
ax.set_xlabel('Month', fontsize = 12)
ax.set_ylabel('Passenger Count', fontsize = 12)
```

### Year and Passenger Count
    Conclusion: Passengers increased gradually as year pass by.
    
    Why? Assumptions
        1. The flight station become more popular.
        2. The flight services improved.
        3. Better flight prices.
        4. Places that the flight offered expanded.
