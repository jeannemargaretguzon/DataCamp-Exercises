```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
```

# PENGUIN DATASET FROM SEABORN EXAMPLE DATASETS

```python
# Importing penguin.csv example dataset 
penguin_data = sns.load_dataset('penguins')
```

```python
penguin_data.head()
```

```python
penguin_data.shape
```

```python
penguin_data.info()
```

```python
print(penguin_data['species'].unique())
print(penguin_data['sex'].unique())
print(penguin_data['island'].unique())
```

```python
# Change the object columns as Categorical Data
penguin_data['species'] = pd.Categorical(penguin_data.species)
penguin_data['sex'] = pd.Categorical(penguin_data.sex)
penguin_data['island'] = pd.Categorical(penguin_data.island)
```

```python
penguin_data.info()
```

**Penguin Population and Gender Per Species** | Stacked Bar

```python
sns.set(style = "ticks")

# Grouping Penguin Sex
penguin_sex = penguin_data.groupby(by = 'sex').size()

fig_2, ax2 = plt.subplots(figsize=(20, 10))

ax2.bar(penguin_data.species, penguin_sex['Male'], color = '#90CAF9', label = 'Male')
ax2.bar(penguin_data.species, penguin_sex['Female'], bottom = penguin_sex['Male'], color = '#EF5350', label = 'Female')

ax2.set_title('Fig 2: Penguin Population and Gender Per Species', fontweight = 'bold', fontsize = 20) 
ax2.set_xlabel('Species', fontsize = 15)
ax2.set_ylabel('Population by Gender', fontsize = 15)
ax.legend()
```

**Species and Body Mass of a Penguin** | Barplot/Column

```python
sns.set_style('whitegrid')

fig_1 = plt.figure(edgecolor = 'black', figsize = (15, 8))
ax1 = sns.barplot(y = 'species', x = 'body_mass_g', data = penguin_data, orient = 'h', ci = None, color ='#90CAF9')
ax1.set_title('Fig 1: Species and Body Mass of a Penguin', fontweight = 'bold', fontsize = 20)   
ax1.set_xlabel('Species', fontsize = 12, fontweight = 'demibold')
ax1.set_ylabel('Body Mass', fontsize = 12, fontweight = 'demibold')
```

```python
sns.set_style('whitegrid')

fig_3 = plt.figure(figsize = (10, 5.75))

ax3 = sns.histplot(data = penguin_data, x = "body_mass_g",  color = '#90CAF9')
ax3.set_title('Fig 3: Body Mass of a Penguin', fontweight = 'bold', fontsize = 20)   
ax3.set_xlabel('Body Mass (in g)', fontsize = 12, fontweight = 'demibold')
ax3.set_ylabel('Count', fontsize = 12, fontweight = 'demibold')
```

Most of the penguins weighed 4000 grams. The Adelie Penguins are penguins that contain the average body mass. Gentoo Penguins are the heaviest penguin among the three.

**Species and Body Parts of a Penguin**

```python
sns.set_style('whitegrid')

fig, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))

sns.barplot(ax = axes[0], x = 'species', y = 'flipper_length_mm', data = penguin_data, ci = None, color ='#FFEB3B')
axes[0].set_title('Which Penguin Species has longer flipper?', fontweight = 'demibold', fontsize = 17)
axes[0].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[0].set_ylabel('Flipper Length (in mm)', fontsize = 12)

sns.barplot(ax = axes[1], x = 'species', y = 'bill_length_mm', data = penguin_data, ci = None, color ='#FFEB3B')
axes[1].set_title('Which Penguin Species has longer bill?', fontweight = 'demibold', fontsize = 17)
axes[1].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[1].set_ylabel('Bill Length (in mm)', fontsize = 12)

sns.barplot(ax = axes[2], x = 'species', y = 'bill_depth_mm', data = penguin_data, ci = None, color ='#FFEB3B')
axes[2].set_title('Which Penguin Species has larger bill?', fontweight = 'demibold', fontsize = 17)
axes[2].set_xlabel('Body Mass (in g)', fontsize = 12)
axes[2].set_xlabel('Bill Depth (in mm)', fontsize = 12)
```

Gentoo Penguins have the longest flipper.
Chinstrap Penguins have the longest bill.
Adelie Penguins and Chinstrap Penguins both have the largest bill.

```python
sns.set(style = "whitegrid")

fig, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))
fig.suptitle('Relation of the Penguin Body Parts with their Body Mass', fontweight = 'bold', fontsize = '20')

# Flipper Length
sns.scatterplot(ax = axes[0], data = penguin_data, x = 'body_mass_g', y = 'flipper_length_mm')
axes[0].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

# Bill Depth 
sns.scatterplot(ax = axes[1], data = penguin_data, x = 'body_mass_g', y = 'bill_length_mm')
axes[1].set(xlabel = 'Body Mass', ylabel = 'Bill Length')

# Bill Length
sns.scatterplot(ax = axes[2], data = penguin_data, x = 'body_mass_g', y = 'bill_depth_mm')
axes[2].set(xlabel = 'Body Mass', ylabel = 'Bill Depth')
```

Longer Flipper and Bill adds on weight that why Gentoo Penguins have highest body mass followed by the Chinstrap Penguins. The Adelie is the lowest among the two category. Although the Adelie and Chinstrap have the same bill depth, it don't add weight that much.
