```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
```

# PENGUIN DATASET FROM SEABORN EXAMPLE DATASETS

```python
# Importing penguin.csv example dataset 
penguin_data = sns.load_dataset('penguins')
```

```python
penguin_data.head()
```

```python
penguin_data.shape
```

```python
penguin_data.info()
```

```python
print(penguin_data['species'].unique())
print(penguin_data['sex'].unique())
print(penguin_data['island'].unique())
```

```python
# Change the object columns as Categorical Data
penguin_data['species'] = pd.Categorical(penguin_data.species)
penguin_data['sex'] = pd.Categorical(penguin_data.sex)
penguin_data['island'] = pd.Categorical(penguin_data.island)
```

```python
penguin_data.info()
```

**Species and Body Mass of a Penguin** | Barplot/Column

```python
# Styling the graph
sns.set_style('whitegrid')

fig_1 = plt.figure(edgecolor = 'black', figsize = (15, 8))
ax1 = sns.barplot(x = 'species', y = 'body_mass_g', data = penguin_data, ci = None, color ='#90CAF9')
ax1.set_title('Fig 1: Species and Body Mass of a Penguin', fontweight = 'bold', fontsize = 20)   
ax1.set_xlabel('Body Mass', fontsize = 12, fontweight = 'demibold')
ax1.set_ylabel('Species', fontsize = 12, fontweight = 'demibold')
```

**Species and Flipper Length of a Penguin** | Horizontal Barplot/Columns

```python
sns.set_style('ticks')

fig_2 = plt.figure(figsize = (15,8))

ax2 = sns.barplot(x = 'flipper_length_mm', y = 'species' , data = penguin_data, orient = 'h', ci = None, color ='#9E9E9E')
ax2.set_title('Fig 2: Which Penguin Species has longer flipper?', fontweight = 'bold', fontsize = 20)
ax2.set_xlabel('Body Mass (in g)', fontweight = 'demibold', fontsize = 12)
ax2.set_ylabel('Flipper Length (in mm)', fontweight = 'demibold', fontsize = 12)
```

**Species and Body Parts of a Penguin** | Scatterplot

```python
sns.set(style = "whitegrid")

fig, axes = plt.subplots(1, 3, sharex = True, figsize = (20, 5.75))
fig.suptitle('Relation of the Penguin Body Parts with their Body Mass', fontweight = 'bold', fontsize = '20')

# Flipper Length
sns.scatterplot(ax = axes[0], data = penguin_data, x = 'body_mass_g', y = 'flipper_length_mm')
axes[0].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

# Bill Depth 
sns.scatterplot(ax = axes[1], data = penguin_data, x = 'body_mass_g', y = 'bill_length_mm')
axes[1].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

# Bill Length
sns.scatterplot(ax = axes[2], data = penguin_data, x = 'body_mass_g', y = 'bill_depth_mm')
axes[2].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')
```

```python
sns.set(style = "whitegrid")

fig, axes = plt.subplots(3, 3, sharex = True, figsize = (20, 5.75))
fig.suptitle('Relation of the Penguin Body Parts with their Body Mass', fontweight = 'bold', fontsize = '20')

# Flipper Length
sns.scatterplot(ax = axes[0], data = penguin_data.species, x = 'Adelie', y = 'flipper_length_mm')
axes[0].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

# Bill Depth 
sns.scatterplot(ax = axes[1], data = penguin_data, x = 'body_mass_g', y = 'bill_length_mm')
axes[1].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

# Bill Length
sns.scatterplot(ax = axes[2], data = penguin_data, x = 'body_mass_g', y = 'bill_depth_mm')
axes[2].set(xlabel = 'Body Mass', ylabel = 'Flipper Length')

ax3 = sns.boxplot(x = 'species', y = 'body_mass_g', data = penguin_data)
```

**Penguin Population and Gender Per Species** | Stacked Bar

```python
sns.set(style = "whitegrid")

# Grouping Penguin Sex
penguin_sex = penguin_data.groupby(by = 'sex').size()

fig, ax = plt.subplots(figsize=(20, 10))

ax.bar(penguin_data.species, penguin_sex['Male'], color = '#29B6F6', label = 'Male')
ax.bar(penguin_data.species, penguin_sex['Female'], bottom = penguin_sex['Male'], color = '#EF5350', label = 'Female')

ax.set_title('Penguin Population and Gender Per Species', fontsize = 15)
ax.set_xlabel('Species', fontsize = 15)
ax.set_ylabel('Population by Gender', fontsize = 15)
ax.legend()
```
